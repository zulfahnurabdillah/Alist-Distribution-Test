```python
columns_to_keep = {
        'Sales Record Number': 'TransactionID',
        'Paid On Date': 'SaleDate',
        'Item Title': 'ItemTitle',
        'Sold For': 'SalePrice',
        'Sold Via Promoted Listings': 'SalesChannel',
        'Ship To Country': 'Country',
        'Quantity': 'Quantity'
    }
```

```python
df_clean = df[columns_to_keep.keys()].rename(columns=columns_to_keep)
```

```python
columns_to_fill = ['TransactionID', 'SaleDate', 'Country'] 
```

```python
df_clean[columns_to_fill] = df_clean[columns_to_fill].fillna(method='ffill')
```

```python
df_clean.dropna(subset=['SalePrice', 'ItemTitle'], inplace=True)
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.drop_duplicates(subset='TransactionID', inplace=True)
```

```python
df_clean.head(5)
```

```python
df_clean['SaleDate'] = pd.to_datetime(df_clean['SaleDate'], format='%b-%d-%y')
```

```python
df_clean['SalePrice'] = df_clean['SalePrice'].astype(str).str.replace(r'[$,]', '', regex=True)
df_clean['SalePrice'] = pd.to_numeric(df_clean['SalePrice'], errors='coerce')
```

```python
df_clean['Quantity'] = pd.to_numeric(df_clean['Quantity'], errors='coerce').fillna(1).astype(int)
```

```python
df_clean['SalesChannel'] = df_clean['SalesChannel'].map({'Yes': 'Promoted', 'No': 'Organic'})
```

```python
df_clean['ItemTitle'].unique
```

```python
brand_map = {
        'chanel': 'Chanel',
        'hermes': 'Hermès',
        'hermès': 'Hermès',
        'louis vuitton': 'Louis Vuitton',
        'rolex': 'Rolex',
        'gucci': 'Gucci',
        'prada': 'Prada',
        'dior': 'Dior',
        'cartier': 'Cartier',
        'goyard': 'Goyard',
        'fendi': 'Fendi'
    }
```

```python
category_map = {
        'Handbag': ['bag', 'clutch', 'birkin', 'kelly', 'speedy', 'neverfull', 'woc', 'pochette', 'tote', 'handbag', 'constance'],
        'Watch': ['watch', 'datejust', 'submariner', 'daytona', 'oyster'],
        'Jewelry': ['necklace', 'bracelet', 'earrings', 'ring', 'jewelry', 'brooch'],
        'Shoes': ['sneakers', 'shoes', 'boots', 'heels', 'sandals', 'loafers'],
        'Accessories': ['earmuffs', 'scarf', 'belt', 'wallet', 'card holder', 'tie', 'sunglasses']
    }
```

```python
def extract_brand(title):
        title_low = str(title).lower()
        for keyword, brand_name in brand_map.items():
            if keyword in title_low:
                return brand_name
        return 'Other'
```

```python
def extract_category(title):
        title_low = str(title).lower()
        for cat_name, keywords in category_map.items():
            for keyword in keywords:
                if keyword in title_low:
                    return cat_name
        return 'Other'
```

```python
df_clean['Brand'] = df_clean['ItemTitle'].apply(extract_brand)
df_clean['Category'] = df_clean['ItemTitle'].apply(extract_category)
```

```python
df_clean.head()
```

```python
df_clean['SaleYear'] = df_clean['SaleDate'].dt.year
df_clean['SaleMonth'] = df_clean['SaleDate'].dt.strftime('%B')
```

```python
df_clean.to_csv('clean_data.csv', index = False)
```
